```python
#---- EXTRACT LAT, LNG and RSSI VALUES FROM txt dataset

# The dataset used here is available on: https://github.com/helderbetiol/LoRaWAN-IC2019
def get_txt_values(filename):
    
    # list of all rows of buildings_settings
    file_rows = []

    # open file in a read mode
    file = open(filename, 'r')

    for line in file.readlines():
        
        # create a str without \n and ","
        line_str = line.strip().split("\t")  
        file_rows.append(line_str)
    
    # remove colunm names
    file_rows.pop(0)
    
    return file_rows

txt_rows = get_txt_values('table-rssi-2805')
```

```python
# ---- UNICAMP DYNAMIC PLOT MAP & MARKERS OF POSITIONS THAT PACKET WERE RECEIVE

# create folium map
fmap = folium.Map(width= 700,height=700,location=[-22.822617, -47.067], zoom_start=15, tiles='OpenStreetMap')

# add markers
for row in txt_rows:
  lat, lng, rssi = row
  folium.Marker(
      location=[float(lat), float(lng)],
      popup="Lat: " + lat + ", Lng: " + lng + ", RSSI: " + rssi,
      icon=folium.Icon(color="red", icon="info-sign"),
  ).add_to(fmap)

fmap.add_child(folium.LatLngPopup())

fmap
```
